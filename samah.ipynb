```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the data set to include only data from the most recent year
latest_year = df['year'].max()
df_latest = df[df['year'] == latest_year]

# Calculate the total CO2 emissions for each country in the most recent year
total_emissions = df_latest.groupby('country_name')['co2'].sum()

# Select the top 5 countries with the highest total CO2 emissions
top_5_countries = total_emissions.nlargest(5).index

# Create a stacked bar chart for each of the top 5 countries, showing CO2 emissions by sector over time
for country in top_5_countries:
    # Filter the data set to include only data for the current country
    df_filtered = df[df['country_name'] == country]
    
    # Calculate the total CO2 emissions for each year
    total_emissions_by_year = df_filtered.groupby('year')['co2'].sum()
    
    # Calculate the percentage of emissions contributed by each sector for each year
    pct_by_year = df_filtered.groupby(['year', 'sector_name'])['co2'].sum() / total_emissions_by_year * 100
    
    # Create a stacked bar chart
    fig, ax = plt.subplots(figsize=(10, 6))
    pct_by_year.unstack().plot(kind='bar', stacked=True, ax=ax)
    ax.set_title(f'{country} CO2 Emissions by Sector ({latest_year})', fontweight='bold', fontsize=14)
    ax.set_xlabel('Year', fontweight='bold', fontsize=12)
    ax.set_ylabel('Percentage of Total CO2 Emissions', fontweight='bold', fontsize=12)
    ax.set_ylim(top=100)
    ax.yaxis.set_major_formatter('{x:.0f}%')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Pivot the dataframe to create a table where each row represents a country, and each column represents a year
pivoted_df = df.pivot(index='country_name', columns='year', values='co2')

# Calculate the percentage change in per capita CO2 emissions for each country between 1960 and 2011
pct_change = (pivoted_df[2011] - pivoted_df[1960]) / pivoted_df[1960] * 100

# Create a bar chart showing the percentage change in per capita CO2 emissions for all countries
fig, ax = plt.subplots(figsize=(10, 6))
pct_change.sort_values().plot(kind='bar', ax=ax, color=['green' if x < 0 else 'red' for x in pct_change])
ax.set_title('Percentage Change in Per Capita CO2 Emissions by Country (1960-2011)', fontweight='bold', fontsize=14)
ax.set_xlabel('Country', fontweight='bold', fontsize=12)
ax.set_ylabel('Percentage Change', fontweight='bold', fontsize=12)
plt.show()
```
